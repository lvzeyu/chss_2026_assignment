## 課題2

pandasでDataフォルダの```state-abbrevs.csv```、```state-areas.csv```、```state-population.csv```を読み込む。

これらのデータを使って、**2010年**アメリカの各州の人口密度を計算しよう。

````python
state_abbrevs_path = Path("Data/state-abbrevs.csv")
df_abbrevs = pd.read_csv(state_abbrevs_path)
df_abbrevs
````

````python
state_areas_path = Path("Data/state-areas.csv")
df_areas = pd.read_csv(state_areas_path)
df_areas
````

````python
state_population_path = Path("Data/state-population.csv")
df_pop = pd.read_csv(state_population_path)
df_pop
````

````python
#各州の名前を略称にする
df_pop2 = pd.merge(df_pop, df_abbrevs,
                left_on="state/region", right_on="abbreviation", how="left")
df_pop2
````

````python
#各州の2010年のtotalの人口のデータを作る
df_2010_total = df_pop2[(df_pop2["year"]==2010) & (df_pop2["ages"]=="total")]
df_2010_total
````

````python
#面積データと州のフルネームをキーにして結合
df_merged = pd.merge(df_2010_total, df_areas, on="state", how="left")
df_merged
````

````python
# state名,面積がNaNの行を除外
df_clean = df_merged.dropna(subset=["state", "area (sq. mi)"])
df_clean
````

````python
# 人口密度を計算
df_clean["density"] = df_clean["population"] / df_clean["area (sq. mi)"]
df_clean
````

## 課題3

```state/region```、```ages```と```years```でデータフレームをグループして、各グループで$\frac{population}{area (sq. mi)}$の形で人口密度を計算してください。

````python
#全人口データに面積データを結合
df_full = pd.merge(df_pop2, df_areas, on="state", how="left")
````

````python
#プエルトリコとUSJの行を消す
df_full = df_full.dropna(subset=["state", "area (sq. mi)"])
df_full
````

````python
#state/region, ages, year でグループ化し、各グループで人口密度を計算
def add_density(group):
    group["density"] = group["population"] / group["area (sq. mi)"]
    return group

df_density = df_full.groupby(["state/region", "ages", "year"]).apply(add_density)
df_density
````
