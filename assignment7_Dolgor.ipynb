## 課題2

pandasでDataフォルダの```state-abbrevs.csv```、```state-areas.csv```、```state-population.csv```を読み込む。

これらのデータを使って、**2010年**アメリカの各州の人口密度を計算しよう。

Task 2. Use pandas to read the following files from the Data folder: 
state-abbrevs.csv
state-areas.csv
state-population.csv
Using these datasets, calculate the population density of each U.S. state in 2010.

````python
import pandas as pd
abbrevs = pd.read_csv("state-abbrevs.csv")
areas = pd.read_csv("state-areas.csv")
population = pd.read_csv("state-population.csv")
````

````python
population = pd.merge(
    population,
    abbrevs,
    left_on="state/region", #The left_on parameter specifies the column in the left DataFrame (population),
    right_on="abbreviation", #specifies the column in the right DataFrame (abbrevs) to join on.
    how="left" #joins the two DataFrames based on the specified columns, keeping all rows from the left DataFrame (population) and adding matching rows from the right DataFrame (abbrevs).
) 
````

````python
population.head()
````

````python
population_2010 = population[
    (population["year"] == 2010) &
    (population["ages"] == "total") 
] ## Use Boolean indexing to filter the DataFrame. And only rows satisfying both conditions are returned.
````

````python
population_2010.head()
````

````python
population_2010["density"] = (
    population_2010["population"] /
    population_2010["area (sq. mi)"]
)
density = population_2010[["state", "density"]] #selecting only the "state" and "density" columns from the population_2010 DataFrame and storing them in a new DataFrame called density
print (density)
````

## 課題3

```state/region```、```ages```と```years```でデータフレームをグループして、各グループで$\frac{population}{area (sq. mi)}$の形で人口密度を計算してください。

Task 3. Group the dataframe by:
state/region
ages
year
Then calculate population density for each group in the requested format.

````python
grouped_population = population.groupby(
    ["state/region", "ages", "year"]
)["population"].sum()
````

````python
print(grouped_population)
````

````python
grouped_population = grouped_population.reset_index() #converting the grouped_population Series back into a DataFrame, with the groupby keys as columns and the aggregated values as a separate column. 
print(grouped_population.head())
````

````python
grouped_population = pd.merge(
    grouped_population,
    abbrevs,
    left_on="state/region",
    right_on="abbreviation",
    how="left"
) #adding the state names to the grouped population DataFrame by merging it with the abbrevs DataFrame. 
````

````python
print(grouped_population.head())
````

````python
grouped_population = pd.merge(
    grouped_population,
    areas,
    on="state",
    how="left"
) #adding the area information to the grouped population DataFrame. 
````

````python
print(grouped_population.head())
````

````python
grouped_population["density"] = (
    grouped_population["population"] /
    grouped_population["area (sq. mi)"]
)
````

````python
density = grouped_population[
    ["state/region", "ages", "year", "density"]
]

print(density)
````
