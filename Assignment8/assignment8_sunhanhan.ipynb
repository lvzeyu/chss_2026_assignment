## 課題1

以下の図を再現してください

- 年齢の分布を、異なる階級幅を使って、4つのサブプロットをわけて描画する
   - 階級幅を適当で決めてよい
- 各サブプロットには、
    - x軸ラベル、y軸ラベル、サブプロット番号(a-d)を指定する
    - 水平方向のグリッド線を追加する
    - ヒストグラムの色を`lightblue`に変更する

```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/lvzeyu/css_tohoku/master/css_tohoku/draft/Data/titanic.csv")

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

bins_list = [5, 10, 15, 20]
labels = ["(a)", "(b)", "(c)", "(d)"]

for ax, bins, label in zip(axes.flatten(), bins_list, labels):

    ax.hist(df["age"].dropna(),
            bins=bins,
            color="lightblue",
            edgecolor="black")

    ax.set_xlabel("Age")
    ax.set_ylabel("Count")

    ax.grid(axis="y")

    ax.text(
        0.03,
        0.95,
        label,
        transform=ax.transAxes,
        fontsize=13,
        va="top"
    )

plt.tight_layout()
plt.show()
```

## 課題2

以下の図を再現してください

- 男性乗客と女性乗客の年齢の分布をそれぞれ描画する
- 全乗客の年齢分布を示す灰色ヒストグラムも描画する
- 図のタイトル、x軸ラベル、y軸ラベル、凡例を設定する

```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/lvzeyu/css_tohoku/master/css_tohoku/draft/Data/titanic.csv")

male = df[df["sex"] == "male"]["age"].dropna()
female = df[df["sex"] == "female"]["age"].dropna()
all_age = df["age"].dropna()

plt.figure(figsize=(8,5))

# 全体乘客
plt.hist(
    all_age,
    bins=20,
    color="lightgray",
    label="All",
    edgecolor="black"
)

# 男性
plt.hist(
    male,
    bins=20,
    alpha=0.7,
    color="mediumpurple",
    label="Male"
)

# 女性
plt.hist(
    female,
    bins=20,
    alpha=0.7,
    color="mediumseagreen",
    label="Female"
)

plt.title("Age Distribution of Titanic Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.legend()

plt.show()
```
