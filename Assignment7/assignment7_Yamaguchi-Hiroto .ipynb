## 課題2

pandasでDataフォルダの```state-abbrevs.csv```、```state-areas.csv```、```state-population.csv```を読み込む。

これらのデータを使って、**2010年**アメリカの各州の人口密度を計算しよう。

````python
abbrevs    = pd.read_csv("Data/state-abbrevs.csv")
areas      = pd.read_csv("Data/state-areas.csv")
population = pd.read_csv("Data/state-population.csv")



data_2010 = population[
    (population["year"] == 2010) &
    (population["ages"] == "total")
    ]

data_2010 = data_2010.merge(abbrevs, left_on="state/region", right_on="abbreviation")
data_2010 = data_2010.merge(areas, on="state")

data_2010["density"] = data_2010["population"] / data_2010["area (sq. mi)"]

data_2010[["state", "population", "area (sq. mi)", "density"]]
````

## 課題3

```state/region```、```ages```と```years```でデータフレームをグループして、各グループで$\frac{population}{area (sq. mi)}$の形で人口密度を計算してください。

````python
merged_data = population.merge(abbrevs, left_on="state/region", right_on="abbreviation")
merged_data = merged_data.merge(areas, on="state")

density_data = (
    merged_data.groupby(["state/region", "ages", "year"])
    .agg({"population": "sum","area (sq. mi)": "first"})
    )

density_data["density"] = (
    density_data["population"]/ density_data["area (sq. mi)"]
)

density_data
````
