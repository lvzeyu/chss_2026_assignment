## 課題2

pandasでDataフォルダの```state-abbrevs.csv```、```state-areas.csv```、```state-population.csv```を読み込む。

これらのデータを使って、**2010年**アメリカの各州の人口密度を計算しよう。

````python
# データの読み込み
data_path = Path("state-abbrevs.csv")
abbrevs = pd.read_csv(data_path)
abbrevs

data_path = Path("state-areas.csv")
areas = pd.read_csv(data_path)
areas

data_path = Path("state-population.csv")
population = pd.read_csv(data_path)
population

# 人口密度の計算
# populationとabbrevsを結合する
population = pd.merge(
    population,
    abbrevs,
    left_on = "state/region",
    right_on = "abbreviation"
)

# areaと結合する
data = pd.merge(
    population,
    areas,
    on = "state"
)
# 計算
data["density"] = data["population"] / data["area (sq. mi)"]
data["density"]
````

## 課題3

```state/region```、```ages```と```years```でデータフレームをグループして、各グループで$\frac{population}{area (sq. mi)}$の形で人口密度を計算してください。

````python
# データのグループ化と人口密度の計算
density = (
    data.groupby(["state/region", "ages", "year"])
    .apply(lambda x: x["population"] / x["area (sq. mi)"]) 
)
density
````
