 ## 課題1

任意の二つの数の間に偶数の和を計算する関数を実装しなさい。
例えば、
入力: a = 1, b = 10 → 出力: 30（2 + 4 + 6 + 8 + 10 = 30）

```python
#result = sum_even_numbers_between(0,10)
```

```python
def sum_even_number_between(a,b):
    total = 0
    for x in range(a,b+1):
        if x % 2==0:
            total = total + x
    return total
result = sum_even_number_between(0,10)
print(result)
```

```python
def sum_even_number_between(a,b):
    even_sum = sum([x for x in range(a, b + 1) if x % 2 == 0])
    return even_sum

result = sum_even_number_between(0,10)
print(result)
```

 ## 課題２

ユークリッドの互除法の関数を作成してください。

ユークリッドの互除法とは、2つの整数、aとbの最大公約数を求めるアルゴリズムです。例えば、2485と1162の最大公約数を求めたい場合

$2485 \div 1162 = 2 余り 161$ 

$1162 \div 161 = 7 余り 35$ 

$161 \div 35 = 4 余り 21$ 

$35 \div 21 = 1 余り 14$

$21 \div 14 = 1 余り 7$

$14 \div 7 = 2 余り 0$

割る数と余りを次回の割られる数と割る数にして、次々と計算し、最後に割り切れたら終わりで、最後の割る数が最大公約数です。


- 除数が$0$によるエラーを制御してください。
- 引数に渡す値の大小関係を考慮してください。除数が被除数より大きい場合は、自動的に値を入れ替えるように実装しなさい。

```python
def euc(a,b):
    if a < b:
       a , b = b , a
    if b == 0:
        return "エラーが発生した、除数に0を指定することはできない"
    while b > 0:
        remainder = a % b
        a = b 
        b = remainder
    return a
result = euc (2485,1162)
print(result)
        
```
