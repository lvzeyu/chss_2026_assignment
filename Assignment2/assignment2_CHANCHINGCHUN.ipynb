````python
# 回答
numbers = [1, 2, 3, 2, 4, 3, 5, 4, 6, 57, 57, 6, 18, 7, 9, 8, 10, 9, 0, 0, 10]

target = 100
show_times = 0
for i in numbers:
   if i == target:
      show_times += 1
if show_times > 0:
   print(show_times)
else:
    print("No Exists")
````
