以下の「スイカ割りゲーム」プログラムをクラスの形に整理しなさい。


Organize the following “Watermelon Splitting Game” program into a class.

```python
import random
import math

BOARD_SIZE = 5  # ボードサイズを定義 Define the board size

def generate_position(board_size):
    # スイカとプレイヤーの初期位置をランダムに生成 Randomly generate the initial position of the watermelon and the player
    return (random.randrange(0, board_size), random.randrange(0, board_size))

def determine_move(player_pos, target_pos):
    # スイカに向かってプレイヤーを移動させる方向を決定 Decide the direction in which the player should move toward the watermelon
    current_x, current_y = player_pos
    target_x, target_y = target_pos

    move_x = 1 if current_x < target_x else -1 if current_x > target_x else 0 #on x-axis, move right (+1) or left (-1) depending on the distance between the player and watermelon. If they are at the same position, it is 0. 
    move_y = 1 if current_y < target_y else -1 if current_y > target_y else 0 #it represent the same process on y-axis, movement here should be up (+1) and down (-1). 

    return (move_x, move_y)

def move_player(player_pos, move_x, move_y):
    # プレイヤーの位置を更新 Update the player's position
    new_x = player_pos[0] + move_x
    new_y = player_pos[1] + move_y
    return (new_x, new_y)

def play_game(board_size): 
    suika_pos = generate_position(board_size) #indicates watermelon position
    player_pos = generate_position(board_size) 

    while suika_pos != player_pos:
        move_x, move_y = determine_move(player_pos, suika_pos)
        player_pos = move_player(player_pos, move_x, move_y)
        print("player moved to:", player_pos) #player moved to/player's existing position

    print("Found the watermelon！") #found the watermelon

# ゲームを開始 Start the game
play_game(BOARD_SIZE)
```
