## 課題2

pandasでDataフォルダの```state-abbrevs.csv```、```state-areas.csv```、```state-population.csv```を読み込む。

これらのデータを使って、**2010年**アメリカの各州の人口密度を計算しよう。

````python
from pathlib import Path

data_path_2 = Path("Data/state-abbrevs.csv")
df_2 = pd.read_csv(data_path_2)

data_path_3 = Path("Data/state-areas.csv")
df_3 = pd.read_csv(data_path_3)

data_path_4 = Path("Data/state-population.csv")
df_4 = pd.read_csv(data_path_4)

df_5 = pd.merge(df_4, df_2, 
                    left_on = "state/region", right_on = "abbreviation", how = "outer")

df_6 = pd.merge(df_5, df_3,
                    on = "state", how = "outer")

data2010 = df_6[(df_6["year"] == 2010) & (df_6["ages"] == "total")]
data2010
data2010 = data2010.set_index("state")
pop_density = data2010["population"] / data2010["area (sq. mi)"]
pop_density
````

## 課題3

```state/region```、```ages```と```years```でデータフレームをグループして、各グループで$\frac{population}{area (sq. mi)}$の形で人口密度を計算してください。

````python
density = df_6.groupby(["state/region", "ages", "year"]).apply(
    lambda g: (g["population"] / g["area (sq. mi)"]).iloc[0]
)
density
````
