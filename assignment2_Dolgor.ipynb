Exercise #3. Search for a key in the dictionary.
If the key exists, print its value.
If it does not exist, print "Not Exists"

```python
birth = {
    'Marx': 1818,
    'Weber': 1864,
    'Durkheim': 1858,
    'Parsons': 1902,
    'Merton': 1911
}

value_0 = "Parsons"
value_1 = "Giddens"
if value_0 in birth:
    print(birth[value_0])
else:
    print("Not Exists")

if value_1 in birth:
    print(birth[value_1])
else:
    print("Not Exists")
```
