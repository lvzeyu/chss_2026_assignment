## 課題2

pandasでDataフォルダの```state-abbrevs.csv```、```state-areas.csv```、```state-population.csv```を読み込む。

これらのデータを使って、**2010年**アメリカの各州の人口密度を計算しよう。

````python
import pandas as pd

df2 = pd.read_csv(r"C:\Users\ma_12\Downloads\state-abbrevs.csv")
df3 = pd.read_csv(r"C:\Users\ma_12\Downloads\state-areas.csv")
df4 =pd.read_csv(r"C:\Users\ma_12\Downloads\state-population.csv")

df4_r = df4[(df4["ages"] == "total") & (df4["year"] == 2010)]
merge1 = pd.merge(df2, df3, how = "left", on="state")
merge2 = pd.merge(merge1, df4_r, how = "right", left_on="abbreviation", right_on="state/region")
merge2
merge2["population_density"] = merge2["population"] / merge2["area (sq. mi)"]
merge2
````

## 課題3

```state/region```、```ages```と```years```でデータフレームをグループして、各グループで$\frac{population}{area (sq. mi)}$の形で人口密度を計算してください。

````python
merge3 = pd.merge(merge1, df4, how = "right", left_on="abbreviation", right_on="state/region")
merge3["population_density"] = merge3["population"] / merge3["area (sq. mi)"]

merge3.groupby(["state", "ages", "year"])["population_density"].sum()
````
