## 課題2

pandasでDataフォルダの```state-abbrevs.csv```、```state-areas.csv```、```state-population.csv```を読み込む。

これらのデータを使って、**2010年**アメリカの各州の人口密度を計算しよう。

````python
import pandas as pd

# データの読み込み
abbrevs = pd.read_csv("C:/Users/User/chss_2026_assignment/state-abbrevs.csv")
areas = pd.read_csv("C:/Users/User/chss_2026_assignment/state-areas.csv")
population = pd.read_csv("C:/Users/User/chss_2026_assignment/state-population.csv")

# 2010年・全年齢(total)のみ抽出
pop2010 = population[
    (population["year"] == 2010) &
    (population["ages"] == "total")
]

# 州名を追加
pop2010 = pd.merge(
    pop2010,
    abbrevs,
    left_on="state/region",
    right_on="abbreviation",
    how="left"
)

# 面積データを追加
pop2010 = pd.merge(
    pop2010,
    areas,
    on="state",
    how="left"
)

# 人口密度を計算
pop2010["population_density"] = (
    pop2010["population"] / pop2010["area (sq. mi)"]
)

# 確認
pop2010[["state", "population", "area (sq. mi)", "population_density"]].head()
````

## 課題3

```state/region```、```ages```と```years```でデータフレームをグループして、各グループで$\frac{population}{area (sq. mi)}$の形で人口密度を計算してください。

````python
import pandas as pd

# データの読み込み
abbrevs = pd.read_csv("C:/Users/User/chss_2026_assignment/state-abbrevs.csv")
areas = pd.read_csv("C:/Users/User/chss_2026_assignment/state-areas.csv")
population = pd.read_csv("C:/Users/User/chss_2026_assignment/state-population.csv")

# 州名を追加
df = pd.merge(
    population,
    abbrevs,
    left_on="state/region",
    right_on="abbreviation",
    how="left"
)

# 面積データを追加
df = pd.merge(
    df,
    areas,
    on="state",
    how="left"
)

# state/region・ages・yearでグループ化して人口密度を計算
density = (
    df.groupby(["state/region", "ages", "year"])
      .apply(lambda x: x["population"].sum() / x["area (sq. mi)"].iloc[0])
      .reset_index(name="population_density")
)

# 確認
print(density.head())
````
