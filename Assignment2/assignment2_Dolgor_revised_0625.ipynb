# 課題1

与えられたリスト内重複な値を削除してください。

````python
numbers = [1, 2, 3, 2, 4, 3, 5, 4, 6, 57, 57, 6, 18, 7, 9, 8, 10, 9, 0, 0, 10]
set_numbers = set(numbers) #to remove duplicates and get unique values
print(set_numbers)
````

````python
unique_numbers= list(set(numbers)) #means to convert the set back to a list
unique_numbers
````

````python
unique_numbers = [] #to create an empty list to store unique numbers
for num in numbers:
    if num not in unique_numbers:
        unique_numbers.append(num)

print(unique_numbers)
````

# 課題2

与えられたリスト内の指定した値を検索し、値の出現回数を```print```してください。見つからない場合は```Not Exists```をprintしてください。

```value_0```と```value_１```を実行した結果を確認してください。

````python

numbers = [1, 2, 3, 2, 4, 3, 5, 4, 6, 57, 57, 6, 18, 7, 9, 8, 10, 9, 0, 0, 10]
value_0= 0
value_1= 100
if value_0 in numbers:
    print(numbers.count(value_0))
else:
    print("Not Exists")
if value_1 in numbers:
    print(numbers.count(value_1))
else:
    print("Not Exists")
````

# 課題3

与えられた辞書内の指定したに一致するキーを検索し、そのキーに対応する値を```print```してください。見つからない場合は```Not Exists```を```print```してください。

```value_0```と```value_１```を実行した結果を確認してください。

````python
birth={'Marx': 1818,
       'Weber': 1864,
       'Durkheim': 1858,
       'Parsons': 1902,
       'Merton': 1911}

value_0= "Parsons"
value_1= "Giddens"

if value_0 in birth:
    print(birth[value_0])
else:
    print("Not Exists")
if value_1 in birth:
    print(birth[value_1])
else:
    print("Not Exists")
````
