## 課題2

pandasでDataフォルダの```state-abbrevs.csv```、```state-areas.csv```、```state-population.csv```を読み込む。

これらのデータを使って、**2010年**アメリカの各州の人口密度を計算しよう。

````python
import pandas as pd

abbrevs = pd.read_csv("Data/state-abbrevs.csv")
areas = pd.read_csv("Data/state-areas.csv")
population = pd.read_csv("Data/state-population.csv")

merged = pd.merge(
    population,
    abbrevs,
    left_on="state/region",
    right_on="abbreviation",
    how="left"
)

merged = pd.merge(
    merged,
    areas,
    on="state",
    how="left"
)

data_2010 = merged[
    (merged["year"] == 2010) &
    (merged["ages"] == "total")
].copy()

data_2010 = data_2010.dropna(
    subset=["state", "area (sq. mi)"]
)

data_2010["population_density"] = (
    data_2010["population"] /
    data_2010["area (sq. mi)"]
)

data_2010[
    ["state", "population", "area (sq. mi)", "population_density"]
].sort_values(
    "population_density",
    ascending=False
)
````

## 課題3

```state/region```、```ages```と```years```でデータフレームをグループして、各グループで$\frac{population}{area (sq. mi)}$の形で人口密度を計算してください。

````python
grouped = pd.read_csv("Data/grouped.csv")

def calculate_density(group):
    return (
        group["population"].sum()
        / group["area (sq. mi)"].mean()
    )

population_density = (
    grouped
    .groupby(["state/region", "ages", "year"])
    .apply(calculate_density)
    .reset_index(name="population_density")
)

population_density
````
