# 課題2

与えられたリスト内の指定した値を検索し、値の出現回数を```print```してください。見つからない場合は```Not Exists```をprintしてください。

```value_0```と```value_１```を実行した結果を確認してください。

````python

numbers = [1, 2, 3, 2, 4, 3, 5, 4, 6, 57, 57, 6, 18, 7, 9, 8, 10, 9, 0, 0, 10]
value_0= 0
value_1= 100
````

````python
#回答
count_0=0
for num in numbers:
    if num == value_0:
        count_0 =count_0 +1

if count_0>0:
    print(count_0)
else:
    print ("Not Exists")
````

````python
count_1=0
for num in numbers:
    if num == value_1:
        count_1 =count_1 +1

if count_1>0:
    print(count_1)
else:
    print ("Not Exists")
````

# 課題3

与えられた辞書内の指定したに一致するキーを検索し、そのキーに対応する値を```print```してください。見つからない場合は```Not Exists```を```print```してください。

```value_0```と```value_１```を実行した結果を確認してください。

````python
birth={'Marx': 1818,
       'Weber': 1864,
       'Durkheim': 1858,
       'Parsons': 1902,
       'Merton': 1911}

value_0= "Parsons"
value_1= "Giddens"
````
