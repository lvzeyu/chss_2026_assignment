## 課題2

pandasでDataフォルダの```state-abbrevs.csv```、```state-areas.csv```、```state-population.csv```を読み込む。

これらのデータを使って、**2010年**アメリカの各州の人口密度を計算しよう。

````python
import pandas as pd

abbrevs = pd.read_csv("/Users/otanikazuki/Downloads/state-abbrevs.csv")
areas = pd.read_csv("/Users/otanikazuki/Downloads/state-areas.csv")
population = pd.read_csv("/Users/otanikazuki/Downloads/state-population.csv")

pop = pd.merge(population,abbrevs,how="outer",left_on="state/region",right_on="abbreviation")
pop = pd.merge(pop,areas,how="left",on="state")
pop_2010 = pop[pop["year"] == 2010]

density = pop_2010.groupby(["state/region", "ages", "year"])["population"].sum() / pop_2010.groupby(["state/region", "ages", "year"])["area (sq. mi)"].first()

density

````

## 課題3

```state/region```、```ages```と```years```でデータフレームをグループして、各グループで$\frac{population}{area (sq. mi)}$の形で人口密度を計算してください。

````python
density = pop.groupby(["state/region", "ages", "year"]).apply(lambda x: x["population"].sum() / x["area (sq. mi)"].iloc[0])

density
````
